## Introducción

Un análisis de regresión es una técnica estadística para investigar y modelar la relación entre una variable (variable respuesta o dependiente) y otra u otras variables (variables explicativas o independientes).

Los campos de aplicación pueden ser ingeniería industrial, ciencias sociales, economía, ciencias biológicas, entre muchos otros. El uso del análisis de regresión puede ser:

* Descripción de datos.
* Estimación de parámetros.
* Predicción y estimación.
* Control.

Los procedimientos generales para hacer un análisis de regresión son:

* Obtener los datos.
* Graficar y obtener estadísticas descriptivas.
* Ajustar el modelo.
* Medir que tan bueno es el ajuste del modelo.
* Usar el modelo (predicción, estimación, descripción).

Imagen2.png

Hay tres métodos básicos para la recolección de datos:

* Estudios retrospectivos basados en datos históricos: se basa en usar los datos durante un periodo de tiempo determinado.
* Estudios observacionales: se basa en usar datos observados no experimentales, no se controlan las variables de interés.
* Diseños experimentales: se basan en datos experimentales y las variables son controladas.

Ejemplos en donde puede aplicarse el modelo de regresión lineal:
* El peso (variable dependiente) depende de la estatura (variable independiente).
* El consumo de una persona (variable dependiente) depende de su ingreso (variable independiente).
* La presión arterial de una persona (variable dependiente) depende de su peso (variable independiente).
* La gasolina consumida por un automóvil (variable dependiente) depende de la distancia recorrida por éste (variable independiente).

## Breve historia

El término regresión fue introducido por Francis Galton en su libro *Natural inheritance* (1889), partiendo de los análisis estadísticos de Karl Pearson. Su trabajo se centró en la descripción de los rasgos físicos de los descendientes (variable A) a partir de los de sus padres (variable B). Estudiando la altura de padres e hijos a partir de más de mil registros de grupos familiares, se llegó a la conclusión de que los padres muy altos tenían una tendencia
a tener hijos que heredaban parte de esta altura, pero que revelaban también una tendencia a regresar a la media. Galton generalizó esta tendencia bajo la **“ley de la regresión universal”**: “Cada peculiaridad en un hombre es compartida por sus descendientes, pero en media, en un grado menor.”

![Imagen2](https://images-na.ssl-images-amazon.com/images/I/51KlU6dZoYL._SY344_BO1,204,203,200_QL70_ML2_.jpg)

El término regresión se utilizó por primer a vez en el estudio de variables antropométricas: al comparar la estatura de padres e hijos, resultó que los hijos cuyos padres tenían una estatura muy superior al valor medio tendían a igualarse a éste, mientras que aquellos cuyos padres eran muy bajos tendían a reducir su diferencia respecto a la estatura media; es decir, “regresaban” al promedio.

La primer forma de regresión lineal documentada fue por medio del método de los mínimos cuadrados, el cual fue publicado por Legendre en 1805 y por Gauss en 1809. El término “mínimos cuadrados” proviene de la descripción dada por Legendre *moindres carrés*. Sin embargo, Gauss aseguró que conocía dicho método desde 1795.

Tanto Legendre como Gauss aplicaron el método para determinar, a partir de observaciones astronómicas, las órbitas de cuerpos alrededor del sol. En 1821, Gauss publicó un trabajo en dónde desarrollaba de manera más profunda el método de los mínimos cuadrados, y en dónde se incluía una versión del teorema de Gauss-Márkov.


## Planteamiento del algoritmo


### Regresión Lineal Simple

El modelo de regresión lineal simple es un modelo con sólo una variable explicativa $x$ (o independiente) que está relacionada con una variable respuesta $y$ (o dependiente) a través de una línea recta.

Imagen1.png

El modelo de regresión lineal se puede escribir como

* $Y = \beta_0 + \beta_1 X  $ (modelo determinista) 
* $Y = \beta_0 + \beta_1 X + \varepsilon$ (modelo estocástico poblacional) 
* $y_i = \beta_0 + \beta_1 x_i + \varepsilon_i $ (modelo estocástico muestral)
para $i = 1, \ldots , n$, donde


- $Y$ es la variable respuesta (o variable independiente).
- $X$ es la variable explicativa (o variable dependiente).
- $\beta_0$ es una constante desconocida (intercepto).
- $\beta_1$ es una constante desconocida (pendiente).
- $\varepsilon$ es un error aleatorio desconocido.

El modelo determinista plantea el modelo que proponemos ajustar. En el modelo estocástico poblacional el término de error $\varepsilon$ se introduce porque al observar $X$ y $Y$ se pueden cometer errores de medición, además de que se pueden haber eliminado algunas variables que se consideraron sin importancia, las cuales causan fluctuaciones en la respuesta. En el modelo estocástico muestral se considera que se tiene una muestra aleatoria de tamaño $n$:
$(x_1, y_1),(x_2, y_2), . . . ,(x_n, y_n)$.

La estimación de los parámetros $\beta_0$ y $\beta_1$ por el método de mínimos cuadrados está dado por:



$$ \hat{\beta}_0 = \displaystyle \sum_{i=1}^{n} [\frac{1}{n}-\frac{(X_{i}-\bar{X})}{S_{xx}}\bar{X}]Y_{i} $$

$$ \hat{\beta}_1 = \displaystyle \sum_{i=1}^{n} \frac{(X_{i}-\bar{X})}{S_{xx}}Y_{i}$$

donde $S_{xx}= \displaystyle \sum_{i=1}^{n}(X_{i}-\bar{X})^{2}$






### Regresión Lineal Múltiple

La mayoría de los fenómenos reales son multicausales, por esta razón, un modelo de regresión más acorde a estudios reales es el modelo de regresión lineal múltiple, que es la generalización del modelo simple. En este caso supondremos que la variable respuesta $Y$ puede explicarse a través de un conjunto de $k$ covariables $X_1, \ldots , X_k$.

El modelo de regresión lineal múltiple se escribe como:

$$Y_i = \beta_0 + \beta_1 X_{1i} + \beta_2 X_{2i} + \cdots + \beta_k X_{ki} + \varepsilon_{i},$$

para $i = 1, \ldots , n$, donde para cada $i$-ésimo individuo

* $Y_i$ es la variable respuesta,
* $X_{ij}$ es el valor observado de la covariable $j$,
* $\varepsilon_i$ es el término de error.


La siguiente figura muestra un conjunto de datos para $k=2$ covariables:

Imagen3.png

La estimación por el método de mínimos cuadrados de los parámetros de regresión está dada por:

$$\hat{\beta}=(X^{'}_{p \times n}X_{n \times p})^{-1}X^{'}_{p \times n}Y_{n \times 1}$$

donde 

$X_{n \times p}=\left(\begin{array}{ccccc}
          1 &   X_{11}  &  X_{12} & \cdots & X_{1k}\\
        1 &   X_{21}  &  X_{22} & \cdots & X_{2k}\\ 
        \vdots & \vdots  &  \vdots  &  \ddots & \vdots \\
        1 &   X_{n1}  &  X_{n2} & \cdots & X_{nk}\\
    \end{array}\right)$ y $Y_{n \times 1}= \left(\begin{array}{c}
                                             Y_{1} \\
                                             Y_{2} \\
                                             \vdots \\
                                             Y_{n}\\
                                            \end{array}\right)$

### Regresión Lineal Simple

El objetivo de un estudio realizado en la Universidad Nacional Autónoma de México es modelar por medio de una regresión lineal la relación entre el IQ y el nivel de plomo en el cuerpo de 99 estudiantes de la Facultad de Ciencias. 

```python
#Cargar los datos 
datos = pd.read_csv("C:/Users/Valeria/Documents/UNAM/Modelos NP Y REG/Regresion/Tarea5/plomo.csv")
#veamos cuantas dimensiones y registros contiene
datos.shape
```

```python
datos.head()
```

```python
#Medidas descriptivas
datos.describe()
```

```python
datos.keys()
```

```python
#Ajuste del modelo
x = datos['Plomo '].values.reshape(-1,1)
y = datos['IQ'].values.reshape(-1,1)

regr = linear_model.LinearRegression()

#Entrenar modelo
regr.fit(x,y)

#Predicciones 
y_pred = regr.predict(x)

# Veamos los coeficienetes obtenidos, En nuestro caso, serán la Tangente
print('La estimación de 𝛽1:', regr.coef_)
# Este es el valor donde corta el eje Y (en X=0)
print('La estimación de 𝛽0:', regr.intercept_)
# Error Cuadrado Medio
print("ECM: %.2f" % mean_squared_error(y, y_pred))
# Puntaje de Varianza. El mejor puntaje es un 1.0
print('Porcentaje de varianza: %.2f' % r2_score(y, y_pred))
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.xlabel('Nivel de Plomo')
plt.ylabel('IQ')
plt.show()
```

## Conclusiones

Por lo tanto, el modelo está dado por $y=95.09803 - 0.13296x $. Esto indica que el intercepto no cruza el origen, es decir, que en caso de que no exista plomo en el cuerpo, el nivel de IQ será de alrededor de 95. Por otro lado, muestra que por cada unidad de plomo que aumente en el cuerpo, el IQ del sujeto va a disminuir en un 13.29 %, en este caso, existe 
una relación inversa entre el plomo y el IQ (esto se puede inferir por el signo negativo de $\beta_1$).
